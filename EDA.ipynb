## Analysis

```python
def categorical_biv(data,tar,cat):
    
    data=data[[cat,tar]]
    
    
    print("Contingency Table:-")
    table=pd.crosstab(data[tar],data[cat])
    print(table)
    
    obs_freq= np.array([table.iloc[:,0].values, table.iloc[:,1].values])
    
    from scipy.stats import chi2_contingency
    
    chi,p,dof,expected = chi2_contingency(obs_freq)
    
    if p<0.05:
        sig=True
    else:
        sig=False
    
    fig,ax=plt.subplots(1,2,figsize=(14,7))
    
    # Grouped Countplot
    sns.set(palette='RdBu')
    sns.countplot(x=cat, hue=tar, data=data,ax=ax[0])
    ax[0].legend(loc='upper right')
    fig.suptitle(f"p-value = {round(p,3)}\n difference significant? = {sig}\n")
    
    ax1 = data.groupby(cat)[tar].value_counts(normalize=True).unstack()
    ax1.plot(kind='bar',ax=ax[1], stacked='True')
    
    ax[1].legend(loc='upper right')
    plt.show()
```

```python
categorical_biv(train,"Response","Gender")
```

```python
categorical_biv(train,"Response","Driving_License")
```

```python
categorical_biv(train,"Response","Previously_Insured")
```

```python
categorical_biv(train,"Response","Vehicle_Age")
```

```python
categorical_biv(train,"Response","Vehicle_Damage")
```

```python
categorical_biv(train,"Response","Region_Code")
```

```python
categorical_biv(train,"Response","Policy_Sales_Channel")
```

**Takeaways-**

- Males are indeed more likely to get a vehicle insurance
- People who had a DL before are more likely to get it again
- People who don't have DL before, are more likely to get one
- People with >2 years old Vehicle are more likely to get a car insurance
- People with previously Damaged Vehicles are more likely to get a car insurance
- Region Code doesn't have much effect on Response
- Policy Sales Channel doesn't have much effect on Resonse

```python
def numerical_Response(var,data):
    
    
    fig,ax=plt.subplots(1,2,figsize=(14,7))
    
    sns.barplot(y=var,x="Response",data=train,ax=ax[0])
    sns.violinplot(x="Response", y=var,
                    data=train,ax=ax[1]) 
    
    
    
    x1=data[data.Response==1][var]
    x2=data[data.Response==0][var]
    
    xm1=x1.mean()
    xm2=x2.mean()
    sd1=x1.std()
    sd2=x2.std()
    n1=len(x1)
    n2=len(x2)

    sd= (sd1*sd1/n1)+(sd2*sd2/n2)

    t_stat=(xm2-xm1)/sd**0.5

    from scipy.stats import t

    p_val=round(2*(1-t.cdf(abs(t_stat),n1+n2-2)),5)
    
    print("P-Value is "+str(p_val))
```

```python
numerical_Response("Age",train)
```

```python
numerical_Response("Annual_Premium",train)
```

```python
numerical_Response("Vintage",train)
```

**Takeaways-**
- Average age of getting a Insurance is significantly higher than that of not getting
- People with slightly higher premiums are more likely to get a car insurance 
- Vintage doesn't affect Insurance at all
